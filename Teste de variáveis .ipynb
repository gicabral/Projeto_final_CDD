```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fam = pd.read_excel('family.xlsx')
fam.head()
```

```python
fam = fam.loc[:, ['geo', 2015]]
fam = fam.rename(columns = {'geo': "Country", 2015:'Family workers 2015'})

fam_nam=fam.dropna()

fam_idx=fam_nam.set_index("Country")
fam_idx.head()
```

```python
dataset = pd.read_csv('hapiness_score_2015.csv')
dataset.head()
```

```python
h = dataset.loc[:,]
data_idx=h.set_index("Country")
data_idx.head()
```

```python
dfam = fam_idx.join(data_idx, how='inner')

ter = dfam.loc[:,['Family workers 2015', 'Happiness Score']]

ter.head()
```

```python
ter.corr()
```

Verificamos que a variável Family Workers não tem muita relação com Happiness Score, então não vamos considerá-la.
